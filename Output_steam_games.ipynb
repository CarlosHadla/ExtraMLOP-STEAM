convertir precios gratuitos a 0 y asigno valor negativo a los Nulos

```python
# función que convierte un valor a float y devolverá 0 si falla
def convert_to_float(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return 0.0

# Aplica la función a la columna 'price' del DataFrame
df_games['price'] = df_games['price'].apply(lambda x: convert_to_float(x))
df_games['price'] = df_games['price'].fillna(-10)
```

Nuevos dataframes para generos y specs para usar en ML

```python
df_games = df_games.dropna(subset=['id'])
df_games['genres']=df_games['genres'].fillna('['']')
df_games['specs']=df_games['specs'].fillna('['']')
```

```python
# Crear una copia del DataFrame original
genres_games_df = df_games[['id','genres']].copy()
# Convertir el contenido de la columna 'genres' a listas de Python
genres_games_df['genres'] = genres_games_df['genres'].apply(ast.literal_eval)
# Obtener una lista de todos los géneros únicos en el DataFrame
unique_genres = set()
for genres in genres_games_df['genres']:
    if isinstance(genres, list):
        unique_genres.update(genres)
# Crear columnas dummy para cada género único
for genre in unique_genres:
    genres_games_df[genre] = genres_games_df['genres'].apply(lambda x: 1 if genre in x else 0)
genres_games_df=genres_games_df.drop(['genres'],axis=1)
```

```python
genres_games_df.head()
```

```python
# Crear una copia del DataFrame original
specs_games_df = df_games[['id','specs']].copy()
# Convertir el contenido de la columna 'specs' a listas de Python
specs_games_df['specs'] = specs_games_df['specs'].apply(ast.literal_eval)
# Obtener una lista de todos los géneros únicos en el DataFrame
unique_specs = set()
for specs in specs_games_df['specs']:
    if isinstance(specs, list):
        unique_specs.update(specs)
# Crear columnas dummy para cada género único
for genre in unique_specs:
    specs_games_df[genre] = specs_games_df['specs'].apply(lambda x: 1 if genre in x else 0)
specs_games_df=specs_games_df.drop(['specs'],axis=1)
```

```python
specs_games_df.head()
```

reduccion de dimensionalidad

```python
# Sumar las columnas y crear un DataFrame
onlygenres=genres_games_df.drop(['id'],axis=1)
genres_sum_df = onlygenres.sum().reset_index()
genres_sum_df.columns = ['Genre', 'Count']
# Convertir la columna "Count" a valores numéricos
genres_sum_df['Count'] = pd.to_numeric(genres_sum_df['Count'], errors='coerce')
# Ordenar el DataFrame por la columna "Count" de forma descendente
sorted_genres_df = genres_sum_df.sort_values(by='Count', ascending=False)
# Obtener los 20 géneros más comunes y sus recuentos
top_20_genres = sorted_genres_df.head(20)
# Calcular el porcentaje de aparición de cada género
total_count = top_20_genres['Count'].sum()
top_20_genres['Percentage'] = (top_20_genres['Count'] / total_count) * 100
# Imprimir los resultados
print(top_20_genres)
```

```python
# Sumar las columnas y crear un DataFrame
onlyspecs=specs_games_df.drop(['id'],axis=1)
specs_sum_df = onlyspecs.sum().reset_index()
specs_sum_df.columns = ['Spec', 'Count']
# Convertir la columna "Count" a valores numéricos
specs_sum_df['Count'] = pd.to_numeric(specs_sum_df['Count'], errors='coerce')
# Ordenar el DataFrame por la columna "Count" de forma descendente
sorted_specs_df = specs_sum_df.sort_values(by='Count', ascending=False)
# Obtener los 20 géneros más comunes y sus recuentos
top_20_specs = sorted_specs_df.head(20)
# Calcular el porcentaje de aparición de cada género
total_count = top_20_specs['Count'].sum()
top_20_specs['Percentage'] = (top_20_specs['Count'] / total_count) * 100
# Imprimir los resultados
print(top_20_specs)
```
